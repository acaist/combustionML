```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_hdf('./data/tables_of_fgm.H5')
df.columns
```

```python
# Plotly
import plotly.offline as pyoff
import plotly.plotly as py
import plotly.tools as tls
```

```python
# df_s=df.sample(frac=0.02)
df_s=df.loc[df['zeta']==list(set(df['zeta']))[0]].sample(frac=0.1)
sp='alpha'
fig_db = {
    'data': [
         {'x': df_s['f'],
         'y': df_s['pv'],
         'z': df_s[sp],
         'type':'scatter3d', 
        'mode': 'markers',
          'marker':{
              'size':1
          }
         }
        
    ],
    'layout': {
        'scene':{
            'xaxis':{'title':'mixture fraction'},
            'yaxis': {'title': "progress variable"},
            'zaxis': {'title': sp}
                 }
    }
}
pyoff.iplot(fig_db, filename='multiple-scatter')
```

```python
df_test=pd.read_hdf('sim_check.H5',key='test')
df_pred=pd.read_hdf('sim_check.H5',key='pred')
```

```python
zeta_level=list(set(df_test['zeta']))
zeta_level.sort()
```

```python
df_t=df_test.loc[df_test['zeta']==zeta_level[1]].sample(frac=0.5)
df_p=df_pred.loc[df_pred['zeta']==zeta_level[1]].sample(frac=0.5)
sp='T'
error=df_p[sp]-df_t[sp]
fig_db = {
    'data': [
        
        {'name':'test',
        'x': df_t['f'],
         'y': df_t['pv'],
         'z': df_t[sp],
         'type':'scatter3d', 
        'mode': 'markers',
          'marker':{
              'size':1
          }
        },
        {'name':'predict',
            'x': df_p['f'],
         'y': df_p['pv'],
         'z': df_p[sp],
         'type':'scatter3d', 
        'mode': 'markers',
          'marker':{
              'size':1
          },
        },
        {'name':'error',
            'x': df_p['f'],
         'y': df_p['pv'],
         'z': error,
         'type':'scatter3d', 
        'mode': 'markers',
          'marker':{
              'size':1
          },
         }       
    ],
    'layout': {
        'scene':{
            'xaxis':{'title':'mixture fraction'},
            'yaxis': {'title': "progress variable"},
            'zaxis': {'title': sp}
                 }
    }
}
py.iplot(fig_db, filename='multiple-scatter')
```
